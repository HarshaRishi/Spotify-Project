```python
# Changing the explicit index to release date and changing the datetime format of the index.

tracks.set_index('release_date', inplace = True)
tracks.index = pd.to_datetime(tracks.index, format = 'mixed')
tracks.head(25)
```

```python
#Converting the duration from milliseconds to seconds.

tracks[['duration']] = tracks[['duration_ms']].apply(lambda x : round(x/1000))
tracks.head()
```

```python
#Dropping the 'duration_ms' column.

tracks.drop('duration_ms', axis = 1, inplace = True)
```

```python
tracks.head()
```

```python
tracks.duration.head()
```

```python
#Correlation map

corr_map = tracks.drop(["key", "mode", "explicit"], axis = 1)
corr_map = corr_map.corr(numeric_only = True, method = "pearson")
```

```python
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_map, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'mako', linewidths = 1, linecolor = 'White')
heatmap.set_title("Correlation Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = tracks.sample(int(0.004*len(tracks)))
```

```python
print(len(sample_df))
```

```python
#Regression Plot

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
#Regression Plot - downward trend

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'maroon').set(title = 'Popularity vs Acousticness Correlation')
```

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates, errors = 'ignore')
years = tracks.dates.dt.year
```

```python
# Histogram plot to understand the trends in number of songs release per year.

sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number of Songs per Year')
```

```python
# Bar plot to understand the trends in average minutes in a song over the years. 

total_dr = tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
pal = sns.color_palette("husl",101)
fig = sns.barplot(x = years, y = total_dr, ax = ax, palette = pal, hue = years, legend = False, err_kws = {'linewidth': 0.0}).set(title = 'Years vs Song Duration (in minutes)')
plt.xticks(rotation = 90)
```

```python
# Line plot visualization for Years vs Duration (in minutes)

total_dr = tracks.duration
fig_dims = (10,5)
sns.set_style(style = "whitegrid")
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = 'Year vs Song Duration (in mins) Line Plot')
plt.xticks(rotation = 60)
```

```python
# Bar Plot - Duration of the songs in different genres

plt.title("Duration of the songs in different genres")
pal = sns.color_palette("husl",27)
sns.barplot(x = 'duration_ms', y = 'genre', data = df_genre, palette = pal, hue = 'genre', legend = False, err_kws = {'linewidth': 0.0})
plt.xlabel("Duration is milliseconds")
plt.ylabel('Genres')
```

```python
# Top 5 Genres by Popularity

sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10,5))
pal = sns.color_palette('flare', 10)
famous = df_genre.sort_values('popularity', ascending = False).head(10)
sns.barplot(x = 'genre', y = 'popularity', data = famous, palette = pal[::-1]).set(title = 'Top 5 Genres by Popularity')
```
